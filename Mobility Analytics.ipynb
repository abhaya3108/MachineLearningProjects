```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn import model_selection, tree, metrics,svm
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge, Lasso, BayesianRidge
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import Normalizer, LabelEncoder
from xgboost.sklearn import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, StackingClassifier
from lightgbm import LGBMClassifier
from math import sqrt
import warnings
warnings.filterwarnings("ignore")

data = pd.read_csv(r"C:\Users\panda4\Documents\Data Science\Online\train_Wc8LBpr.csv")
data
```

```python
data.isnull().sum()
```

```python
data['Type_of_Cab'].fillna('U', inplace=True)
data['Life_Style_Index'].fillna(data['Life_Style_Index'].median(), inplace=True)
data['Customer_Since_Months'].fillna(data['Customer_Since_Months'].median(), inplace=True)
data['Confidence_Life_Style_Index'].fillna('U', inplace=True)
data['Var1'].fillna(data['Var1'].median(), inplace=True)
data
```

```python
columns = ['Type_of_Cab', 'Confidence_Life_Style_Index', 'Destination_Type']
for col in columns:
    encoder = LabelEncoder().fit(data[col])
    data[col] = encoder.transform(data[col])
data
```

```python
data['Gender'] = data['Gender'].map({'Female':0,'Male':1})
```

```python
X = data.drop(['Trip_ID','Surge_Pricing_Type'], axis=1)
X
```

```python
y = pd.DataFrame(data['Surge_Pricing_Type'], columns=['Surge_Pricing_Type'])
y
```

```python
test_size = 0.3
seed = 3
X_train, X_valid, y_train, y_valid = model_selection.train_test_split(X, y, test_size = test_size, random_state = seed)
```

```python
models = []
models.append(('GB', XGBClassifier()))
models.append(('Forest', RandomForestClassifier()))
models.append(('Light', LGBMClassifier(boosting = "gbdt",
    metric="multi_logloss")))
```

```python
model_comp = pd.DataFrame(columns=['Model', 'Accuracy'])

for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    accuracy = accuracy_score(y_valid, y_pred)
    model_comp = model_comp.append([{'Model': name, 'Accuracy': accuracy}])

model_comp
```
