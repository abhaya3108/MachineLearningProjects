```python
import seaborn as sns
data = sns.load_dataset("titanic")
data
```

```python
data = data[['sex','embarked','alone','pclass','survived']]
data.head()
```

```python
import numpy as np
data['sex']=np.where(data['sex']=='male',1,0)
ordinal_label = {k:i for i,k in enumerate(data['embarked'].unique(),0)}
data['embarked']=data['embarked'].map(ordinal_label)
data['alone']=np.where(data['alone']==True,1,0)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data[['sex','embarked','alone','pclass']],
    data['survived'],
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

```python
X_train.isnull().sum()
```

```python
from sklearn.feature_selection import chi2
f_p_values=chi2(X_train,y_train)
```

```python
f_p_values
```

```python
import pandas as pd
p_values = pd.Series(f_p_values[1])
p_values.index = X_train.columns
p_values.sort_index(ascending=False)
```
