### Question 8. What is the percentage prior probability of the occurrence of '1' in the predicted variable?

```python
X = train.drop(['Unnamed: 0', 'collision'],axis=1)
y = train['collision']
```

```python

model = LinearDiscriminantAnalysis()

model.fit(X, y)

```

```python
y_pred = model.predict(X)
```

```python
metrics.classification_report(y_true=y, y_pred=y_pred)
```
