```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv(r"C:\Users\panda4\Documents\Data Science\notebooks\winequality-white.csv", delimiter=';')
data
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values

correlation = data.corr()
fig = plt.subplots(figsize=(10,10))
sns.heatmap(correlation,vmax=1,square=True,annot=True,cmap='Reds')
```

```python
features = data.columns.tolist()[:-1]
features
for feature in features:
    sns.set()
    sns.relplot(data=data, x=feature, y=Y, kind='line', height=4, aspect=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=4)
linreg = LinearRegression()
linreg.fit(x_train,y_train)
y_pred = linreg.predict(x_test)

accuracy = linreg.score(x_test, y_test)
print(accuracy)

rmse_linear = mean_squared_error(y_test, y_pred)
print(rmse_linear)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pydotplus
import seaborn as sns
data = pd.read_csv(r"C:\Users\panda4\Documents\Data Science\notebooks\winequality-white.csv", delimiter=';')
data
```

```python
data.columns
```

```python
y = data['quality']
X = data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
from sklearn import tree, metrics, model_selection, preprocessing
#from Python.display import Image, display
```

```python
X_train,X_test,y_train,y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=4)
X_test.shape
```

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
count_misclassified = (y_test != y_pred).sum()
print(count_misclassified)
accuracy = metrics.accuracy_score(y_test, y_pred)
print(accuracy)
```
