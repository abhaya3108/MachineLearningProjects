# Feature Selection - dropping features which are not relevant

```python
data_step1.drop(['id', 'wkda_created_on', 'estimated_price_3rd_party_provider'], axis=1, inplace=True)
```

# Feature Engineering

```python
data_step1['built_year'] = 2021 - data_step1['built_year']
data_step1['first_quote_wkda'].fillna(data_step1['first_quote_wkda'].median(), inplace=True)
```

```python
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
data_step1['manufacturer']= label_encoder.fit_transform(data_step1['manufacturer'])
data_step1['main_type'] = label_encoder.fit_transform(data_step1['main_type'])
```

```python
y = data_step1['first_quote_wkda']
X = data_step1.drop('first_quote_wkda', axis=1)
```

# Train test split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from xgboost.sklearn import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from lightgbm import LGBMRegressor
```

```python
models = []
models.append(('Linear', LinearRegression()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('GB', XGBRegressor()))
models.append(('Forest', RandomForestRegressor()))
models.append(('Light', LGBMRegressor()))
models.append(('ADA', AdaBoostRegressor()))
```

```python
from math import sqrt

model_comp = pd.DataFrame(columns=['Model', 'RMSE'])

for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = sqrt(mean_squared_error(y_test, y_pred))
    model_comp = model_comp.append([{'Model': name, 'RMSE': rmse}])

model_comp
```
