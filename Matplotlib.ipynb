```python
import pandas as pd
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic
```

```python
#Create a pie chart presenting the male/female proportion
import matplotlib.pyplot as plt
%matplotlib inline
sex = titanic.groupby("sex")["sex"].count()
plt.pie(sex, colors = ["green", "yellow"], labels = ["Female", "Male"], autopct='%1.2f', startangle=90)
plt.title("Male/Female proportion", fontname="Times New Roman Bold", fontweight='bold', fontsize = 20)
plt.show()
```

```python
#Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
titanic.dropna(how = 'all', inplace = True)
colors = {'male':'blue', 'female':'orange'}
plt.figure(figsize = (15,10))
plt.scatter(titanic.fare, titanic.age, c=titanic['sex'].apply(lambda x:colors[x]))
plt.xlabel("Fare Paid")
plt.ylabel("Age")
plt.title("Fare paid and the Age", fontname="Times New Roman Bold", fontweight='bold', fontsize = 20) 
plt.show()
```
